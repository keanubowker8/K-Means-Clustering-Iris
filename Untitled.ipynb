```python
X = whiskey_df[['Body', 'Sweetness', 'Smoky', 'Medicinal', 'Tobacco',
                'Honey', 'Spicy', 'Winey', 'Nutty', 'Malty', 'Fruity', 'Floral', 'Latitude', 'Longitude']].values
```

```python
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(X)
    distortions.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

<blockquote>
    After having performed anaylsis on the features I saw that it was forming 3 or 4 clusters. Now after looking at this graph it gives me confirmation to use 4 as the number of optimal clusters.
    </blockquote>

```python
kmeans = KMeans(n_clusters=4, random_state=1)
kmeans
```

```python
kmodel = kmeans.fit(X)
kmodel
```

```python
plt.subplots(1,figsize=(10,10))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=kmodel.labels_,s=100)
#plt.legend(['Iris-versicolor','Iris-setosa','Iris-virginica'])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
#plt.set_title('Predicted', fontsize=18)
```
